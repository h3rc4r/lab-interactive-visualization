```python
import pandas as pd
import plotly.plotly as py
import cufflinks as cf
from ipywidgets import interact

cf.go_offline()
```

```python
data = pd.read_excel('C:/Users/b4bsas/Documents/lab-interactive-visualization/data/Online Retail.xlsx')
data.head()
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
data_no_uk = data[(data['Country']!='United Kingdom')&(data['InvoiceDate']>='2011-04-01 00:00:00')&(data['InvoiceDate']<='2011-04-30 23:59:59')]
data_no_uk = data_no_uk.groupby(['Country'], as_index=False)['Quantity','Revenue'].sum()
data_no_uk
```

```python
data_no_uk.iplot(kind='bar', x='Country', xTitle='Country',
           yTitle='Total', title='total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
party = data[data.Description=='PARTY BUNTING'][['Quantity','UnitPrice','Country']]
party = party.groupby('Country', as_index=False).mean()
party
```

```python
party.iplot(x='Quantity', y='UnitPrice', categories='Country',
           xTitle='Average Quantity', yTitle='Average Unit Price',
           title='Quantity and unit price average for PARTY BUNTING by country')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
countries = ['EIRE', 'Germany', 'France', 'Netherlands']
h_countries= data[data.Country.isin(countries)][['Quantity','InvoiceNo','Country']]

h_countries = h_countries.pivot_table(values='Quantity',columns='Country',index='InvoiceNo',aggfunc='sum')
h_countries.head()
```

```python
h_countries.iplot(kind='hist' , subplots=True
         ,xTitle='Invoice', yTitle='Quantity',title='Quantity per Invoice')
```

## 5. Create an interactive side-by-side bar chart showing the revenue by country listed below (bars) for each of the products listed below.

```python
product_list = ['JUMBO BAG RED RETROSPOT', 
                'CREAM HANGING HEART T-LIGHT HOLDER',
                'REGENCY CAKESTAND 3 TIER']

country_list = ['EIRE', 'Germany', 'France', 'Netherlands']
bars = data[(data['Country'].isin(country_list))&(data['Description'].isin(product_list))][['Description','Country','Revenue']]
bars = bars.pivot_table(values='Revenue',columns='Country',index='Description',aggfunc='sum')
bars
```

```python

bars.iplot(kind='bar',
         
         xTitle='Product', yTitle='Revenue',title='Revenue by Product')
```
